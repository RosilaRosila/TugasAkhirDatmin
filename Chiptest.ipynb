```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('chip_dataset_fix.xlsx')
df
```

```python
df.columns = ['No','Product','Type','ReleaseDate','RelaseDateFix','ProcessSize','TDP','DieSize','Transsistors','Freq','Foundry','Vendor','FP16','Fp32','FP64']
df
```

```python
df.dtypes
```

```python
df.isna().sum()
```

```python
def_ReleaseDate = "0"
df.ReleaseDate.fillna(def_ReleaseDate, inplace =True)
def_ProcessSize = "0"
df.ProcessSize.fillna(def_ProcessSize, inplace =True)
def_TDP = "0"
df.TDP.fillna(def_TDP, inplace =True)
def_DieSize = "0"
df.DieSize.fillna(def_DieSize, inplace =True)
def_Transsistors = "0"
df.Transsistors.fillna(def_Transsistors, inplace =True)
def_FP16 = "0"
df.FP16.fillna(def_FP16, inplace =True)
def_Fp32 = "0"
df.Fp32.fillna(def_Fp32, inplace =True)
def_FP64 = "0"
df.FP64.fillna(def_FP64, inplace =True)
df
```

```python
df.isna().sum()
```

```python
df
```

```python
for index, row in df.iterrows():
    df.loc[index, 'TahunRelease'] = row['RelaseDateFix'][-4:]
df
```

```python
from datetime import datetime
now = datetime.now()

now.year
for index, row in df.iterrows():
    df.loc[index, 'LamaRelease_Tahun']=now.year - int (row['TahunRelease'])
df
```

```python
d_train = df[['No','Product','Type','ProcessSize','TDP','DieSize','Transsistors','Freq','TahunRelease','LamaRelease_Tahun']].reset_index(drop=True)
d_train
```

```python
for index, row in d_train.iterrows():
    if row['LamaRelease_Tahun'] >=7:
        d_train.loc[index, 'Keterangan_Release'] = 'lama'
    elif row['LamaRelease_Tahun'] >=6:
        d_train.loc[index, 'Keterangan_Release'] = 'sedang'
    else:
        d_train.loc[index, 'Keterangan_Release'] = 'Baru'
d_train
```

```python
target = d_train['Keterangan_Release'].values
target
```

```python
columns = ['No','ProcessSize','TDP','DieSize','Transsistors','Freq']
feature = d_train[list(columns)]
feature
```

```python
#membagi dataset menjadi data test dan training
from sklearn.model_selection import train_test_split
```

```python
independen_train, independen_test, dependen_train, dependen_test = train_test_split(feature, target, test_size=0.3)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score, precision_score, classification_report
from sklearn import tree
#from sklearn.preprocessing import Imputer
from IPython.display import Image
from subprocess import check_call
```

```python
clf = tree.DecisionTreeClassifier(criterion='entropy')
clf = clf.fit(independen_train, dependen_train)
predict = clf.predict(independen_test)
```

```python
with open("Graph_tree.dot", 'w') as f:
    f = tree.export_graphviz(clf, out_file=f, feature_names=columns,
                           filled=True, rounded=True, special_characters=True)
```

```python
import seaborn as sns
f, ax = plt.subplots(figsize=(5,3))
sns.heatmap(confusion_matrix(dependen_test, predict),annot=True, fmt=".0f",ax=ax)
plt.xlabel("predict")
plt.ylabel("actual")
plt.show()
```

```python
print(classification_report(dependen_test, predict))
```

```python
accuracy_score(dependen_test, predict)
```
